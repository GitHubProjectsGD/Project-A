```python
# bin Email_Status to get either 0 or 1
df['conversion'] = df['Email_Status'].apply(lambda x: 0 if x < 2 else 1)   
                                         
```

```python
# total number of conversions out of total number of rows
print('total conversions: %i out of %i' % (df.conversion.sum(), df.shape[0])) 
```

```python
# conversion rate to two decimal places 
print('conversion rate: %0.2f%%' % (df.conversion.sum() / df.shape[0] * 100.0))
```

```python
# the number of conversions per each group of Time Email Sent Category   
pd.DataFrame(
    df.groupby(
        by='Time_Email_sent_Category'
    )['conversion'].sum()
)
```

```python
# The total number per each group of time email sent category - althought conversion is listed, each group includes conversions and non conversions - listed word conversion is default    
pd.DataFrame(
    df.groupby(
        by='Time_Email_sent_Category'
    )['conversion'].count()
)
```

```python
# the % of conversions per each group of time email sent category  
conversions_by_Emails = df.groupby(
    by='Time_Email_sent_Category'
)['conversion'].sum() / df.groupby(
    by='Time_Email_sent_Category'
)['conversion'].count() * 100.0
```

```python
pd.DataFrame(conversions_by_Emails)
```

```python
# number of non conversions and conversions per each group of time email sent. 
conversions_by_Emails = pd.pivot_table(df, values='Email_Status', index='Time_Email_sent_Category', columns='conversion', aggfunc=len)
conversions_by_Emails
```

```python
# changed to say non conversions and conversions  
conversions_by_Emails.columns = ['non_conversions', 'conversions']
```

```python
conversions_by_Emails
```

```python
#Among converters compared to non converters, higher % converted in Time Email sent Category 1 
#Amoung converters compared to non converters, higher % converted in Time Email sent Category 2. 
#Amoung converters compared to non converters, lower % converted in Time Email sent Category 3. 
#Numbers between converters and non converters are similar so times email sent is likely not going to be a strong predictor of conversion
 
conversions_by_Emails.plot(
    kind='pie',
    figsize=(15, 7),
    startangle=90,
    subplots=True,
    autopct=lambda x: '%0.1f%%' % x
)

plt.show()
```

To answer question #2, there 2373 conversions in the dataset, leading to a 3.47% conversion rate.
```python
# Install train test split
from sklearn.model_selection import train_test_split
```

```python
# splitting the data frame - 80% training, 20% testing 
X_train, X_test, y_train, y_test = train_test_split(df[Extra_Credit] , df['conversion'], test_size=0.2)
```

```python
df[Extra_Credit].shape 
```

```python
# checking the Xtrain shape - Total Data frame is 68,353 - 80% is 54682
X_train.shape 
```

```python
# checking the Xtest shape - total data frame is 68,353 - 20% is 13671
X_test.shape
```

```python
# the target variable
y_train.shape 
```

```python
y_train.head (10)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
import statsmodels.formula.api as sm
```

```python
# To build a linear regression model
```

```python
new_model = sm.logit(
    df[Extra_Credit], 
    df['conversion']
)
```

```python
new_model.fit (X_train, y_train)
```

```python
new_model_preds = new_model.predict(X_test)
```

```python
# to check the accuracy of the training data 
accuracy_score(y_train, new_model.predict(X_train))
```

```python
# to check the accuracy of the test data 
accuracy_score (y_test , new_model_preds)
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
# to compute the ROC AUC Score 
in_sample_preds = new_model.predict_proba(X_train)[:,1]
out_sample_preds = new_model.predict_proba(X_test)[:,1]
```

```python
in_sample_fpr, in_sample_tpr, in_sample_thresholds = roc_curve(y_train, in_sample_preds)
out_sample_fpr, out_sample_tpr, out_sample_thresholds = roc_curve(y_test, out_sample_preds)
```

```python
in_sample_roc_auc = auc(in_sample_fpr, in_sample_tpr)
out_sample_roc_auc = auc(out_sample_fpr, out_sample_tpr)

print('In-Sample AUC: %0.4f' % in_sample_roc_auc)
print('Out-Sample AUC: %0.4f' % out_sample_roc_auc)
```

```python
plt.figure(figsize=(10,7))

plt.plot(
    out_sample_fpr, out_sample_tpr, color='darkorange', label='Out-Sample ROC curve (area = %0.4f)' % in_sample_roc_auc
)
plt.plot(
    in_sample_fpr, in_sample_tpr, color='navy', label='In-Sample ROC curve (area = %0.4f)' % out_sample_roc_auc
)
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.grid()
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve')
plt.legend(loc="lower right")

plt.show()
```

To complete Extra Credit #1, I changed the NaN cells in Total Links to 0 and created a new column named Total Links New. I created a data frame with Subject Hotness Score, Word Count and Total Links New, split the data frame to 80% training data and 20% testing data. I built a logistic regression model and checked the accuracy of my target training and testing data. I computed the ROC AUC score and graphed the ROC curve. The gap between the in sample ROC curve and the out of sample ROC curve is small which indicates very little overfitting.
